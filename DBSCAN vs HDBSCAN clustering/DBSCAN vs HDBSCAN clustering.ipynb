```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import hdbscan
import numpy as np
import matplotlib.pyplot as plt

# Create clusters with different densities
centers = [(0, 0), (5, 5), (10, 0)]
cluster_std = [0.2, 1.0, 2.5]  # different densities

X, _ = make_blobs(n_samples=800, centers=centers, cluster_std=cluster_std, random_state=42)

plt.scatter(X[:, 0], X[:, 1], s=10)
plt.title("Variable Density Clusters")
plt.show()
```

```python
# -----------------------------
#  Apply DBSCAN
# -----------------------------
dbscan = DBSCAN(eps=0.5, min_samples=10)
labels_dbscan = dbscan.fit_predict(X)

# -----------------------------
#  Apply HDBSCAN
# -----------------------------
hdb = hdbscan.HDBSCAN(min_cluster_size=10)
labels_hdb = hdb.fit_predict(X)

# -----------------------------
# : Plot results
# -----------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Original Data
axes[0].scatter(X[:, 0], X[:, 1], s=10)
axes[0].set_title("Original Data")

# DBSCAN Result
axes[1].scatter(X[:, 0], X[:, 1], c=labels_dbscan, cmap='rainbow', s=10)
axes[1].set_title("DBSCAN Result")

# HDBSCAN Result
axes[2].scatter(X[:, 0], X[:, 1], c=labels_hdb, cmap='rainbow', s=10)
axes[2].set_title("HDBSCAN Result")

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=500, noise=0.08, random_state=42)

# Add random noise
noise = np.random.uniform(low=-3, high=3, size=(100, 2))
X = np.vstack([X, noise])

plt.scatter(X[:, 0], X[:, 1], s=10)
plt.title("Two Moons with Noise")
plt.show()
```

```python
# -----------------------------
#  Apply DBSCAN
# -----------------------------
dbscan = DBSCAN(eps=0.5, min_samples=10)
labels_dbscan = dbscan.fit_predict(X)

# -----------------------------
#  Apply HDBSCAN
# -----------------------------
hdb = hdbscan.HDBSCAN(min_cluster_size=10)
labels_hdb = hdb.fit_predict(X)

# -----------------------------
# : Plot results
# -----------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Original Data
axes[0].scatter(X[:, 0], X[:, 1], s=10)
axes[0].set_title("Original Data")

# DBSCAN Result
axes[1].scatter(X[:, 0], X[:, 1], c=labels_dbscan, cmap='rainbow', s=10)
axes[1].set_title("DBSCAN Result")

# HDBSCAN Result
axes[2].scatter(X[:, 0], X[:, 1], c=labels_hdb, cmap='rainbow', s=10)
axes[2].set_title("HDBSCAN Result")

plt.tight_layout()
plt.show()
```
